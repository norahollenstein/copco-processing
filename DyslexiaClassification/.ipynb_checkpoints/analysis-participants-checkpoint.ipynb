```python
# import libraries

import numpy as np
import math
import sklearn as sk
import pandas as pd
from scipy import stats
import researchpy as rp
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('../participant_stats.csv')
```

```python
df = df[df.subj != 'P00']
df = df[df.subj != 'P01']
df = df[df.subj != 'P13']
df = df[df.subj != 'P14']
df = df[df.subj != 'P17']
df = df[df.subj != 'P32']
```

```python
dys = df.loc[df['dyslexia'] == 'yes']
non = df.loc[df['dyslexia'] == 'no']
```

```python
non.sort_values(by=['subj'])
```

```python
def welchs_ttest(feature):
    dys_data = dys[feature]
    nondys_data = non[feature]

    score = rp.ttest(dys_data, nondys_data, equal_variances=False)
    t = score[1]['results'][2] 
    p = score[1]['results'][3]
    
    print(score)
    
    print("\nsignifinance testing of feature:", feature)
    print("t-test score:", t)
    print("p-value:", "{0:.15f}".format(p))

    if p < 0.05:
        print("difference between dyslexic and nondyslexic: significant")
    else:
        print("difference between dyslexic and nondyslexic: insignificant")
        
```

```python
welchs_ttest('absolute_reading_time')
```

```python
welchs_ttest('relative_reading_time')
```
