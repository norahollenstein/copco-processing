```python
# import libraries

import numpy as np
import math
import sklearn as sk
import pandas as pd
from scipy import stats
import researchpy as rp
import seaborn as sn
```

```python
lex = pd.read_excel('lex_assess_test.xlsx')
```

```python
lex = lex[(lex.participant != 'P32')]
```

```python
# store groups individually
non_dys = lex.iloc[:10]
dys = lex.iloc[10:]
```

```python
def ttest_score(feature):
    dys_data = dys[feature]
    nondys_data = non_dys[feature]

    score = rp.ttest(dys_data, nondys_data, equal_variances=True)
    t = score[1]['results'][2] 
    p = score[1]['results'][3]
    
    print(score)
    
    print("\nsignifinance testing of feature:", feature)
    print("t-test score:", t)
    print("p-value:", "{0:.15f}".format(p))

    if p < 0.05:
        print("difference between dyslexic and nondyslexic: significant")
    else:
        print("difference between dyslexic and nondyslexic: insignificant")
        
```

```python
ttest_score('reading_score')
```
